when we use `__repr__()` it prints the value which is seen by the python and the python developers want to see.

It doesn't only print value like `__str__()`.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return self.title

    def __repr__(self):
        return f"Book(title='{self.title}', pages={self.pages})"

b = Book("Python OOP", 350)

print(b)
b
```

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book(title='{self.title}', pages={self.pages})" # User-friendly display

    def __repr__(self):
        return f"Book(title='{self.title}', pages={self.pages})" # Debug / inspection

b = Book("Python OOP", 350)

print(b)
b
```
