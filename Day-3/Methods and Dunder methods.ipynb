```python
class Land:
    def __init__(self, name):
        self.value = name # the name (local variable) value is copied into objects instance variable value

    def val(self):
        return f"{self.value} is my name"

A2 = Land("Aayush")
print(A2.val())
```

`__new__()` → creates objects

`__init__()` → initializes objects

`__str__()` → Defines how an object looks when printed.

`__repr__()` → Gives an official, unambiguous representation of the object.

```python
class A:
    def __str__(self):
        return f"is my name"

    def val(self):
        return f"is my name"

obj = A()
print(obj) # __str__() is used for printing the value automatically without needing to get an extra method to print it.
print(obj.val())
```

```python
class User:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"User('{self.name}')"

    def __repr__(self):
        return f"User('{self.name}')"

u = User("Aayush")
print(u) # here __str__() is working
u        # here __repr__() is working
```

```python
class Animal:
    birds = None

    def __new__(cls, *args, **kwargs):
        if cls.birds is None:
            print("Making an Object for birds")
            cls.birds = super().__new__(cls)
            return cls.birds
        else:
            print("Object already created")
            return cls.birds

    def __init__(self, name):
        self.name = name
        print(f"Bird created {self.name}")

A1 = Animal("Aayush")
A2 = Animal("Sparrow")
print(A1 is A2) # class variable is shared among both the objects

```
